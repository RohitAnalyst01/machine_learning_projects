```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs 
```

```python
data = make_blobs(n_samples=200,n_features=2,cluster_std=1.8,random_state=101)
```

```python
sns.scatterplot(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=4)
```

```python
km.fit(data[0])
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
fig, (axis1,axis2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

axis1.set_title('kmeans')
axis1.scatter(data[0][:,0],data[0][:,1],c= km.labels_,cmap='rainbow')

axis2.set_title('original')

axis2.scatter(data[0][:,0],data[0][:,1],c= data[1],cmap='rainbow')
```
