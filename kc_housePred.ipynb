```python
df = df.drop('zipcode',axis=1)
```

```python
df['yr_renovated'].value_counts()
```

```python
df['sqft_basement'].value_counts()
```

```python
X = df[['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','waterfront', 'view', 'condition', 'grade', 'sqft_above','sqft_basement', 'yr_built', 'yr_renovated','sqft_living15', 'sqft_lot15', 'year', 'month']]
y = df['price']
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data = (X_test,y_test),batch_size = 128,epochs=400)
```

```python
model.history.history
```

```python
pd.DataFrame(model.history.history)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
np.sqrt(mean_absolute_error(y_test,predictions))
```

```python
explained_variance_score(y_test,predictions)
```
