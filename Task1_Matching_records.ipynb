#The Approach 
The following steps were followed:
1. Uploading the liberaries (i.e. Pandas and Numpy) and datasets as df1 and df2.
2. So, In order to retrive the from both df1 and df2 we need to outer join the datasets on the primary keys (i.e. Product ID+Order ID).
3. Product ID, Order ID are 2 different columns. Therefore, they have to be merged to form a single column just for the ease of use.
4. Now, In order to find the missing data, .isna() is used to find the missing values within the column.
5. The result from the function gives the dataset with missing values. Then, calling the .shape() gives the number of the missing values.
6. For finding number of missing quantities, .sum() is applied to the sliced dataset.

```python
#importing liberaries
import numpy as np
import pandas as pd
```

```python
# reading the Data from the csv or given file format
df1 = pd.read_csv('Records Matching Task.xlsx - data1.csv')
```

```python
df1
```

```python
#Creating combined ID(Order ID+Product ID)

df1['Uni_id1'] = df1['Order ID']+df1['Product ID']
```

```python
df1
```

```python
df2 = pd.read_csv('Records Matching Task.xlsx - data2.csv')

```

```python
df2
```

```python
#Creating Unique column ID - "Data1_Unique_ID"
df2['Uni_id2'] = df2['Order ID']+df2['Product ID']

df2
```

```python
# Two data sets are merged on the common Unique ID
merged_df = df1.merge(df2, how='outer', left_on=['Uni_id1'], right_on=['Uni_id2'])
```

```python
merged_df
```

**Identifying the Records (Order ID + Product ID combination) present in data1 but missing in data2 (Specify the number of records missing in your answer)**

```python
missing_df2 = merged_df[merged_df['Uni_id2'].isna()]
missing_df2
```

```python
#No. of missing values in df2
missing_df2.shape[0]
```

**Identifying the Records (Order ID + Product ID combination) present in data2 but missing in data1 (Specify the number of records missing in your answer)**

```python
missing_df1 = merged_df[merged_df['Uni_id1'].isna()]
```

```python
missing_df1
```

```python
#No. of missing values in df1
missing_df1.shape[0]
```
