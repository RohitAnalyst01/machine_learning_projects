```python
m = merged_df.groupby('Book-Title').count()['Book-Rating']>=50
books_popular = m[m]
```

```python
real_rating = merged_df[merged_df['Book-Rating'].isin(books_popular)]
```

```python
pivot = real_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot
```

```python
pivot.fillna(0,inplace=True)
```

```python
pivot
```

```python
from scipy.sparse import csr_matrix
```

```python

book_sparse = csr_matrix(pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(pivot.iloc[237, :].values.reshape(1, -1))
```

```python
for i in range(len(suggestions)):
  print(pivot.index[suggestions[i]])
```
